#### getting venue details and addding it to make a new dataframe

```python
def getnearbyvenues(name,lat,lng):
    venuelist=[]
    for n,la,ln in zip(name,lat,lng):
         url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            la, 
            ln, 
            radius, 
            LIMIT)
         results=requests.get(url).json()['response']['groups'][0]['items']
         venuelist.append([(n,la,ln,v['venue']['name'],v['venue']['location']['lat'],v['venue']['location']['lng'],v['venue']['categories'][0]['name']) for v in results])
    nearby_venues=pd.DataFrame([item for venue in venuelist for item in venue])
    nearby_venues.columns=['Neighborhood','latitude','longitude','venue','venue.latitude','venue.longitude','venue.category']
    return(nearby_venues)
        
```

```python
venue_data=getnearbyvenues(df['Neighborhood'],df['Latitude'],df['Longitude'])
```

```python
venue_data.head()
```

#### use get dummies function to make venue categories as colums for k means algo

```python
onehot=pd.get_dummies(venue_data[['venue.category']],prefix='',prefix_sep='')
onehot['Neighborhood']=venue_data['Neighborhood']
```

```python
column=[onehot.columns[-1]] + list(onehot.columns[:-1])
```

```python
onehot=onehot[column]
```

```python
onehot
```

```python
toronto_grouped=onehot.groupby('Neighborhood').mean()
toronto_grouped.reset_index(inplace=True)
```

#####  finding out the most common venues

```python
def return_most_common(row,n):
    row_categories=row.sort_values(ascending=False)
    return row_categories.index.values[0:n]
```

```python
n=10
indicators=['st','nd','rd']
columns=['Neighborhood']
for i in np.arange(n):
    try:
        columns.append(str(i+1) + indicators[i]+'most common')
    except:
        columns.append(str(i+1)+'th most common')
ven=pd.DataFrame(columns=columns)
ven['Neighborhood']=toronto_grouped['Neighborhood']
for i in np.arange(toronto_grouped.shape[0]):
    ven.iloc[i,1:]=return_most_common(toronto_grouped.iloc[i,1:],n)
```

# Clustering

```python
c=4
toronto_cluster=toronto_grouped.drop('Neighborhood',axis=1)
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=c,random_state=0).fit(toronto_cluster)
ven.insert(0,'Cluster',kmeans.labels_)
ven
```

```python

merge_data=pd.merge(df,ven,on=['Neighborhood'])
merge_data
```
